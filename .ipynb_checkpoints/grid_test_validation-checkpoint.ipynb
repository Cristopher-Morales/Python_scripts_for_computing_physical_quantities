```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
x=pd.read_csv('/home/cristopher/Desktop/Documentation/grid_convergence.csv',header=None).values[1:,0] # inputs file
y1=pd.read_csv('/home/cristopher/Desktop/Documentation/grid_convergence.csv',header=None).values[1:,1] # outputs file
y2=pd.read_csv('/home/cristopher/Desktop/Documentation/grid_convergence.csv',header=None).values[1:,2]
# Create Plot
x=np.float_(x)
y1=np.float_(y1)
y2=np.float_(y2)

fig, ax1 = plt.subplots(figsize=(8, 6)) 
  
ax1.set_xlabel('$h=\sqrt{1/N}$')
ax1.set_ylabel('Pressure drop [Pa]', color = 'black', fontsize = 16.0) 
plot_1 = ax1.plot(x, y1, color = 'black',markersize=12, marker ='*', label='Pressure drop') 
ax1.tick_params(axis ='y', labelcolor = 'black')
ax1.set_ylim(75, 121)
# Adding Twin Axes

ax2 = ax1.twinx() 
  
ax2.set_ylabel('Mass fraction $CH_{4}$[-]', color = 'r',fontsize = 16.0) 
plot_2 = ax2.plot(x, y2, color = 'r', markersize=12, marker ='o', label='Mass fraction $CH_{4}$') 
ax2.tick_params(axis ='y', labelcolor = 'r')
ax2.set_ylim(0.356, 0.3581)
# Show plot
ax1.legend(loc='upper right')
ax2.legend(loc='lower right')
fig.suptitle('Pressure drop and mass fraction methane\n grid convergence', fontsize = 20.0) #defining titles
#plt.set_figure(figsize=(14,8))
fig.tight_layout()
#plt.rcParams.update({'font.size': 14})
fig.savefig('/home/cristopher/Pictures/FIGURES_EPS/mesh_refinement.pdf', format='pdf',transparent=False, dpi=1200)
#plt.show()
```
